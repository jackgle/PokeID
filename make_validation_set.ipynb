```python
import os
from random import shuffle
import subprocess

if not os.path.exists('../test/'):
    os.mkdir('../test/')

for i in os.listdir('../data/'):
    files = os.listdir('../data/'+i)
    shuffle(files)
    print(i+':\t'+str(len(files)))
    
    if not os.path.exists('../test/'+i):
        os.mkdir('../test/'+i)
    
    for jj in files[:int(len(files)*0.1)]:
        
        subprocess.run(['mv', '../data/'+i+'/'+jj, '../test/'+i+'/'+jj])
```
